```python
import pandas as pd
import numpy as np
from scipy.stats import zscore
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv('data/data/benin-malanville.csv')
```

```python

df.describe()

# Missing value report
missing_report = df.isna().sum()
missing_percent = (missing_report / len(df)) * 100
missing_df = pd.DataFrame({'Missing': missing_report, '% Missing': missing_percent})
missing_df[missing_df['% Missing'] > 5]
```

```python
columns_to_check = ['GHI', 'DNI', 'DHI', 'ModA', 'ModB', 'WS', 'WSgust']
z_scores = df[columns_to_check].apply(zscore)
outliers = (np.abs(z_scores) > 3)


df_outliers = df[outliers.any(axis=1)]
```

```python

df_clean = df.copy()
for col in columns_to_check:
    median_val = df_clean[col].median()
    df_clean[col] = df_clean[col].fillna(median_val)


df_clean = df_clean.dropna(subset=columns_to_check)
```

```python
df_clean.to_csv('data/benin_clean.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure timestamps are parsed correctly
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df.set_index('Timestamp', inplace=True)
```

```python
# Optionally resample to hourly/daily average to smooth
df_daily = df.resample('D').mean()

plt.figure(figsize=(14, 6))
plt.plot(df_daily.index, df_daily['GHI'], label='GHI')
plt.plot(df_daily.index, df_daily['DNI'], label='DNI')
plt.plot(df_daily.index, df_daily['DHI'], label='DHI')
plt.plot(df_daily.index, df_daily['Tamb'], label='Tamb')
plt.legend()
plt.title('Daily Average of GHI, DNI, DHI, and Tamb')
plt.xlabel('Date')
plt.ylabel('Value')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df['Month'] = df.index.month
monthly_avg = df.groupby('Month')[['GHI', 'DNI', 'DHI', 'Tamb']].mean()

monthly_avg.plot(kind='bar', figsize=(12, 6))
plt.title("Monthly Averages of GHI, DNI, DHI, Tamb")
plt.ylabel("Average Value")
plt.xlabel("Month")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df['Hour'] = df.index.hour
hourly_avg = df.groupby('Hour')[['GHI', 'DNI', 'DHI', 'Tamb']].mean()

hourly_avg.plot(figsize=(12, 6))
plt.title("Hourly Average Trends")
plt.xlabel("Hour of Day")
plt.ylabel("Average Value")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
if 'Cleaning' in df.columns:
    cleaning_impact = df.groupby('Cleaning')[['ModA', 'ModB']].mean()
    
    cleaning_impact.plot(kind='bar', figsize=(8, 5))
    plt.title("Average ModA & ModB Before vs After Cleaning")
    plt.ylabel("Average Sensor Reading")
    plt.xticks(ticks=[0, 1], labels=["Before Cleaning", "After Cleaning"], rotation=0)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
else:
    print("No 'Cleaning' column found in dataset.")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


corr_cols = ['GHI', 'DNI', 'DHI', 'TModA', 'TModB']
corr_matrix = df[corr_cols].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```

```python
scatter_vars = [('WS', 'GHI'), ('WSgust', 'GHI'), ('WD', 'GHI'), ('RH', 'Tamb'), ('RH', 'GHI')]

for x, y in scatter_vars:
    plt.figure(figsize=(6, 4))
    sns.scatterplot(data=df, x=x, y=y, alpha=0.5)
    plt.title(f"{y} vs. {x}")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
